```python
import pandas as pd
import re
```

```python
streams = pd.read_sql_table(table_name="radiometadata", con="sqlite:////home/peter7rom/Projects/python/metadata_parser/metadata.db", index_col="id").reset_index()
streams.head()
```

```python
streams.rename(columns={"id": "Id", "title": "StreamName", "land": "Country", "mainland": "Mainland", "logo": "Logo", "url": "Url"}, inplace=True)
streams
```

```python
import numpy as np
streams['StreamName'].replace('', np.nan, inplace=True)
streams.dropna(subset=["StreamName"], inplace=True)
streams.head()
```

```python
for i in set(list(streams.Country)):
    print(i)
```

```python
import json

data = dict()
countries = sorted(set(list(streams.Country)))
for i in countries:
    data.setdefault(i, []).append(streams.query(f"Country == '{i}'").Id.to_list())
query = streams.query("IsFavouriteStream == 'True'")
data.setdefault("favourite_stream", []).append(query.Id.to_list())
for i, v in data.items():
    data[i] = v[0]
with open("/home/peter7rom/Projects/go/src/github.com/tech7strann1k/online-radio/db/metadata.json", "w") as file:
    json.dump(data, file)
```

```python
streams.sort_values(["Country", "StreamName"], inplace=True)
streams
```

```python
streams.drop_duplicates(subset="StreamName", keep="first", inplace=True)
```

```python
splt_logos = streams.Logo.str.split('/')
splt_logos
```

```python
logos = [i[-1:][0] for i in splt_logos]
logos
streams.Logo = logos
streams.Logo = streams.Logo.replace(to_replace=r"\.\w+", value=".jpg", regex=True)
streams.Logo
```

```python
streams.StreamName = streams.StreamName.replace(to_replace=r"([\d\/\.]+ (FM|MHz|МГц))", value="", regex=True)
```
